```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/atm_transactions.csv')

df.describe()
```

```python
# Extract numerical columns from the DataFrame
numerical_columns = df.describe().columns

# Plot histograms for each numerical column
for column in numerical_columns:
    plt.figure(figsize=(4, 2))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Plot a histogram of ATM's totalBalance values
totalBalanceSum = df.groupby('atmName')['totalBalance'].sum().reset_index()
plt.figure(figsize=(4, 2))
sns.histplot(totalBalanceSum['totalBalance'], kde=True)
plt.title('Distribution of Total Balance by ATM')
plt.xlabel('Total Balance Sum')
plt.ylabel('Frequency')
plt.show()
```

### **Histogram of Highest Single Withdrawal Amount Grouped by Week**

```python
# Group by week and ATM to get the highest single withdrawal amount
df['transactionWeek'] = df['transactionTime'].dt.isocalendar().week
highest_weekly_withdrawal = df.groupby(['atmName', 'transactionWeek'])['totalOutcome'].max().reset_index()

# Plot a histogram of the highest single withdrawal amount grouped by week
plt.figure(figsize=(10, 6))
sns.histplot(highest_weekly_withdrawal['totalOutcome'], kde=True)
plt.title('Highest Single Withdrawal Amount by Week')
plt.xlabel('Highest Single Withdrawal Amount')
plt.ylabel('Frequency')
plt.show()
```

### **Daily Total withdrawal by ATM**

```python
# Convert the transactionTime to datetime
df['transactionTime'] = pd.to_datetime(df['transactionTime'])

# Extract the date from transactionTime
df['transactionDate'] = df['transactionTime'].dt.date

# Group by atmName and transactionDate to get the daily total withdrawals
daily_withdrawals = df.groupby(['atmName', 'transactionDate'])['totalOutcome'].sum().reset_index()

# Plot a histogram of the daily total withdrawals
plt.figure(figsize=(4, 2))
sns.histplot(daily_withdrawals['totalOutcome'], kde=True)
plt.title('Distribution of Daily Total Withdrawals by ATM')
plt.xlabel('Daily Total Withdrawals')
plt.ylabel('Frequency')
plt.show()
```

### **Histogram on Weekly Total Withdrawals Grouped by Day of the Week**

```python
# Group by day of the week and ATM to get weekly total withdrawals
df['transactionDayOfWeek'] = df['transactionTime'].dt.day_name()
weekly_withdrawals = df.groupby(['atmName', 'transactionDayOfWeek'])['totalOutcome'].sum().reset_index()

# Plot a histogram of the weekly total withdrawals grouped by day of the week
plt.figure(figsize=(6, 3))
sns.histplot(weekly_withdrawals['totalOutcome'], kde=True)
plt.title('Distribution of Weekly Total Withdrawals by Day of the Week')
plt.xlabel('Weekly Total Withdrawals')
plt.ylabel('Frequency')
plt.show()
```

### **Distribution of Withdrawals by Day of the Week**

```python
# Group by day of the week and ATM to get weekly total withdrawals
df['transactionDayOfWeek'] = df['transactionTime'].dt.day_name()
weekly_withdrawals = df.groupby(['transactionDayOfWeek'])['totalOutcome'].sum().reset_index()

# Plot a barplot of the weekly total withdrawals grouped by day of the week
plt.figure(figsize=(10, 6))
sns.barplot(x='transactionDayOfWeek', y='totalOutcome', data=weekly_withdrawals, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Weekly Total Withdrawals by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Withdrawals')
plt.show()
```

```python
# Group by ATM to get the total withdrawal amount
withdrawals_per_atm = df.groupby('atmName')['totalOutcome'].sum().reset_index()
withdrawals_per_atm_sorted = withdrawals_per_atm.sort_values(by='totalOutcome', ascending=False)

# Plot a barplot of the ATMs with the highest total withdrawal amount
plt.figure(figsize=(100, 50))
sns.barplot(x='atmName', y='totalOutcome', data=withdrawals_per_atm_sorted)
plt.title('ATM with the Highest Total Withdrawal Amount')
plt.xlabel('ATM')
plt.ylabel('Total Withdrawal Amount')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group by month and ATM to get monthly total withdrawals
df['transactionMonth'] = df['transactionTime'].dt.month
monthly_withdrawals = df.groupby(['atmName', 'transactionMonth'])['totalOutcome'].sum().reset_index()

# Plot a histogram of the monthly total withdrawals grouped by month
plt.figure(figsize=(10, 6))
sns.histplot(monthly_withdrawals['totalOutcome'], kde=True)
plt.title('Distribution of Monthly Total Withdrawals by Month')
plt.xlabel('Monthly Total Withdrawals')
plt.ylabel('Frequency')
plt.show()
```
